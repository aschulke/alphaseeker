```python

import pandas as pd
import numpy as np
from matplotlib.ticker import FuncFormatter
from matplotlib.ticker import FormatStrFormatter
import matplotlib.pyplot as plt
%matplotlib inline


#read the second sheet (sheet_2)
#SPX data cleaned into excel and imported, below reads excel and prints names
file= r'T:\Operations\Personal\Carter E\Python Docs\Composition of Indicies.xlsx'
VWO = pd.read_excel(file,
                    sheetname="VWOdata",
                  parse_cols=[0,1],
                    header=[0,1]
                   )

VWO['Weight'].plot(kind='barh')
plt.legend(['Percentage'])
plt.title('Percentage Weight in VWO')
plt.xlim(0, .3)
plt.show();

file= r'T:\Operations\Personal\Carter E\Python Docs\Composition of Indicies.xlsx'
EEM = pd.read_excel(file,
                    sheetname="EEMdata",
                  parse_cols=[0,1],
                    header=[0,1]
                   )

EEM['Weight'].plot(kind='barh')
plt.legend(['Percentage'])
plt.title('Percentage Weight in EEM')
plt.xlim(0, .3)
plt.show();
```
